```python
from sklearn import preprocessing
encode = preprocessing.LabelEncoder()
titanic['Sex'] = encode.fit_transform(titanic['Sex'].astype(str))
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.linear_model import SGDClassifier
from sklearn.svm import LinearSVC
from sklearn.neighbors import RadiusNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.model_selection import train_test_split
```

```python
res_dict={}
```

```python
def summerize_classification(y_test,y_pred):
    acc = accuracy_score(y_test, y_pred, normalize=True)
    num_acc = accuracy_score(y_test, y_pred, normalize=False)
    
    prec = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    return {'accuracy': acc,
            'precision': prec,
            'recall': recall,
            'accuracy_count':num_acc
    }
```

```python
def build_model(classifier_fn,
               name_of_y_col,
               names_of_x_cols,
               dataset,
               test_frac=0.2):
    X = dataset[names_of_x_cols]
    Y = dataset[name_of_y_col]
    x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
    model = classifier_fn(x_train, y_train)
    y_pred = model.predict(x_test)
    y_pred_train = model.predict(x_train)
    
    train_summary = summerize_classification(y_train,y_pred_train)
    test_summary = summerize_classification(y_test, y_pred)
    
    pred_results = pd.DataFrame({'y_test':y_test,
                                'y_pred':y_pred})
    
    model_crosstab = pd.crosstab(pred_results.y_pred, pred_results.y_test)
    return {'training': train_summary,
            'test': test_summary,
            'confusion_matrix': model_crosstab
    }
```

```python
def compare_results():
    for key in res_dict:
        print("Classification: ", key)
        print()
        print('Training data')
        for score in res_dict[key]['training']:
            print(score, res_dict[key]['training'][score])
        print()
        print('test data')
        for score in res_dict[key]['test']:
            print(score, res_dict[key]['test'][score])
        print()
```

```python
def logistic_fun(x_train, y_train):
    model = LogisticRegression(solver='liblinear')
    model.fit(x_train, y_train)
    return model
```

```python
Features = list(titanic.columns[1:])
```

```python
res_dict['survived - logistic'] = build_model(logistic_fun,
                                                'Survived',
                                                Features,
                                                titanic)
compare_results()
```

```python
def dis(x_train, y_train):
    model = LinearDiscriminantAnalysis(solver='svd')
    model.fit(x_train, y_train)
    return model
```

```python
res_dict['survived - linearDiscriminant_analysis'] = build_model(dis,
                                                'Survived',
                                                Features,
                                                titanic)
compare_results()
```

```python
def sgd(x_train, y_train, max_iter=1000, tol=1e-3):
    model = SGDClassifier(max_iter=max_iter, tol=tol)
    model.fit(x_train, y_train)
    return model
```

```python
res_dict['SVD'] = build_model(sgd,
                                                'Survived',
                                                Features,
                                                titanic)
compare_results()
```

```python
def linear_svc(x_train, y_train,C=1.0,max_iter=1000, tol=1e-3):
    model = LinearSVC(max_iter=max_iter, tol=tol, dual=False)
    model.fit(x_train, y_train)
    return model
```

```python
res_dict['svc'] = build_model(linear_svc,
                                                'Survived',
                                                Features,
                                                titanic)
compare_results()
```

```python
def knn(x_train, y_train,radius=40.0):
    model = RadiusNeighborsClassifier(radius=radius)
    model.fit(x_train, y_train)
    return model
```

```python
res_dict['Knn'] = build_model(knn,
                                                'Survived',
                                                Features,
                                                titanic)
compare_results()
```

```python
def tree(x_train, y_train,max_depth=None, max_features=None):
    model = DecisionTreeClassifier(max_depth=max_depth,max_features=max_features)
    model.fit(x_train, y_train)
    return model
```

```python
res_dict['tree'] = build_model(tree,
                                                'Survived',
                                                Features,
                                                titanic)
compare_results()
```

```python
def naive(x_train, y_train,priors=None):
    model = GaussianNB(priors=priors)
    model.fit(x_train, y_train)
    return model
```

```python
res_dict['naive bayes'] = build_model(naive,
                                                'Survived',
                                                Features,
                                                titanic)
compare_results()
```
